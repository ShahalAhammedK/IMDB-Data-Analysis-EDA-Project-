```python
movies_per_year = sorted_new_clean_df.groupby('year').size()
movies_per_year.columns = ['year','number_of_movies']
movies_per_year.columns
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Assuming 'sorted_new_clean_df' contains the DataFrame and 'year' is the column we need
# Create a series with the number of movies per year
movies_per_year = sorted_new_clean_df.groupby('year').size()

# Convert the series into a DataFrame
movies_per_year_df = movies_per_year.reset_index(name='number_of_movies')

# Set the year range for the slider
min_year = 1900
max_year = 2025

# Set end_year to the maximum year
end_year = max_year  # Fixed as 2025

# Function to plot the chart based on selected years
def plot_movies_per_year(start_year):
    # Filter data within the selected year range
    filtered_data = movies_per_year_df[(movies_per_year_df['year'] >= start_year) & (movies_per_year_df['year'] <= end_year)]

    # Plotting the bar chart
    plt.figure(figsize=(10,6))
    plt.bar(filtered_data['year'], filtered_data['number_of_movies'], color='skyblue')

    # Adding titles and labels
    plt.title(f'Number of Movies per Year ({start_year}-{end_year})', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Number of Movies', fontsize=12)

    # Display the plot
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Create a slider for selecting the start_year between 1900 and 2025
year_slider = widgets.IntSlider(
    value=min_year,  # Default to the min year (1900)
    min=min_year,
    max=max_year,  # Max is set to 2025
    step=1,
    description='Start Year:',
    continuous_update=False
)

# Use the interact function to update the plot based on the slider
interact(plot_movies_per_year, start_year=year_slider)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Assuming 'sorted_new_clean_df' contains the DataFrame with 'year' and 'revenue' columns
# Create a series with the total revenue per year
revenue_each_year = sorted_new_clean_df.groupby('year')['revenue'].sum()

# Function to plot the revenue per year
def plot_revenue_per_year(start_year, end_year):
    filtered_revenue_data = revenue_each_year[(revenue_each_year.index >= start_year) & (revenue_each_year.index <= end_year)]

    # Plotting the line chart for revenue
    plt.figure(figsize=(10, 6))
    plt.plot(filtered_revenue_data.index, filtered_revenue_data.values, marker='o', color='g', markersize=7, linewidth=3, linestyle='-')

    # Adding titles and labels
    plt.title(f'Total Revenue per Year ({start_year} - {end_year})', fontsize=16)
    plt.xlabel('Year', fontsize=16)
    plt.ylabel('Total Revenue (in $)', fontsize=16)
    plt.grid(True)

    # Display the plot
    plt.tight_layout()
    plt.show()

# Set the year range for the slider
min_year = revenue_each_year.index.min()
max_year = revenue_each_year.index.max()

# Create the slider widgets for selecting the start and end years
start_year_slider = widgets.IntSlider(min=min_year, max=max_year, step=1, value=min_year, description='Start Year:')
end_year_slider = widgets.IntSlider(min=min_year, max=max_year, step=1, value=max_year, description='End Year:')

# Use interact with both sliders
interact(plot_revenue_per_year, start_year=start_year_slider, end_year=end_year_slider)
```

```python
# Ensure that 'release_date' is in datetime format
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Now extract the year, month, day, and quarter
df['year'] = df['release_date'].dt.year
df['month'] = df['release_date'].dt.month
df['day'] = df['release_date'].dt.day
df['quarter'] = df['release_date'].dt.quarter
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML

# Assuming 'new_clean_df' is your IMDb dataset
data = df

# Ensure necessary columns exist and handle any missing columns
if 'profit' not in data.columns:
    data['profit'] = data['revenue'] - data['budget']

# Extract unique values for dropdown options
unique_years = sorted(data['year'].dropna().unique().tolist())
unique_genres = sorted(data['genres'].dropna().unique().tolist())
unique_countries = sorted(data['production_countries'].dropna().unique().tolist())
unique_languages = sorted(data['original_language'].dropna().unique().tolist())

# Create widgets for user input
text_title = widgets.Text(description='Title:')
dropdown_year = widgets.Dropdown(options=[None] + unique_years, description='Year:')
dropdown_genre = widgets.Dropdown(options=[None] + unique_genres, description='Genre:')
dropdown_country = widgets.Dropdown(options=[None] + unique_countries, description='Country:')
dropdown_language = widgets.Dropdown(options=[None] + unique_languages, description='Language:')

button = widgets.Button(description='Search')

# Function to handle button click
def handle_button_click(button):
    # Get input values
    selected_title = text_title.value.lower() if text_title.value else None
    selected_year = dropdown_year.value
    selected_genre = dropdown_genre.value
    selected_country = dropdown_country.value
    selected_language = dropdown_language.value

    # Apply filters to the dataset
    filtered_data = data.copy()
    if selected_title:
        filtered_data = filtered_data[filtered_data['Title'].str.lower().str.contains(selected_title)]
    if selected_year:
        filtered_data = filtered_data[filtered_data['year'] == selected_year]
    if selected_genre:
        filtered_data = filtered_data[filtered_data['genres'].str.contains(selected_genre, na=False)]
    if selected_country:
        filtered_data = filtered_data[filtered_data['production_countries'].str.contains(selected_country, na=False)]
    if selected_language:
        filtered_data = filtered_data[filtered_data['original_language'] == selected_language]

    # Sort by vote_average and popularity
    sorted_data = filtered_data.sort_values(['vote_average', 'popularity'], ascending=False)

    # Select relevant columns and display the top results
    columns_to_display = ['title', 'year', 'revenue', 'genres', 'original_language', 'vote_average', 'popularity', 'profit']
    top_results = sorted_data[columns_to_display].head(10)

    display(HTML(top_results.to_html(index=False)))

# Connect the button click event to the handler function
button.on_click(handle_button_click)

# Display widgets
display(text_title)
display(dropdown_year)
display(dropdown_genre)
display(dropdown_country)
display(dropdown_language)
display(button)
```
